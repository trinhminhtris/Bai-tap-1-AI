```python
import pandas as pd
dataframe = pd.DataFrame()
dataframe['Name'] = ['Lemon', 'Apple', 'Banana', 'Orange']
dataframe['Price'] = [5, 8, 7, 9]
dataframe['Lisence'] = [False, True, True, False]

# create row data
new = pd.Series(['Strawberry', 10, True], index = ['Name', 'Price', 'Lisence'])
# append row
dataframe.append(new, ignore_index = True)
```

```python
# append a column
dataframe['Location'] = ['USA', 'VN', 'SIN', 'Canada']
dataframe
```

```python
# delete a row
dataframe.drop(2, axis = 0).head(10)
```

```python
# delete a column
dataframe.drop('Price', axis = 1).head(10)
```

```python
dataframe['Price'].replace(9, 10).head(5)
```

```python
# rename columns
dataframe.rename(columns = {'Price':'Cost'}).head()
```
